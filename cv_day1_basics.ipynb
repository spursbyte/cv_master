```python
import argparse
import cv2
import sys
sys.path.append("../")
from google.colab.patches import cv2_imshow
import numpy as np

```

```python
args={
    'image':'..//content/blog2.png'
}
image=cv2.imread(args['image'])
print('width: %d pixels'% (image.shape[1]))
print('height: %d pixels'% (image.shape[0]))
print('channels: %d channels'% (image.shape[2]))
cv2_imshow(image)
(h,w)=image.shape[:2]



```

ACCESS PIXELS


```python
(B,G,R)=image[0,0]
print("Pixel at (0,0) - Red: {r}, Green: {g}, Blue: {b}".format(r=R, g=G, b=B))
#slicing of image
(cx,cy)=(w//2,h//2)
top_left=image[0:cy,0:cx]
cv2_imshow(top_left)
top_right=image[0:cy,cx:w]
cv2_imshow(top_right)



```

```python
bottom_left= image[cy:h,0:cx]
cv2_imshow(bottom_left)
```

```python
bottom_right=image[cy:h,cx:w]
cv2_imshow(bottom_right)
```

```python
#drawing line, circle, and rectangle over a canvas
canvas=np.zeros((300,300,3),dtype='uint8')
#draw green line
green=(0,255,0)
cv2.line(canvas,(0,0),(300,300),green)
cv2_imshow(canvas)
```

```python
#red line
red=(0,0,255)
cv2.line(canvas,(0,300),(300,0),red)
cv2_imshow(canvas)
```

```python
#for rectange, we pass the top-left and bottom-right corner coordinates to the function
cv2.rectangle(canvas,(10,20),(230,250),green)
cv2_imshow(canvas)
```

```python

#reset canvas and get center of canvas for drawing circle
canvas = np.zeros((300,300,3), dtype='uint8')
(centerX, centerY) = (canvas.shape[1]//2, canvas.shape[0]//2)
white = (255,255,255)
for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white)
cv2_imshow(canvas)
```
